```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc = {'figure.figsize':(12,6)})
```

```python
df = pd.read_csv("SuperstoreSales_main.csv", encoding = 'ISO- 8859-1', delimiter = ";")
returns_df = pd.read_csv("SuperstoreSales_returns.csv", encoding = 'ISO- 8859-1', delimiter = ";")

# i600_df = df.loc[(df['Product Name'] == 'Okidata Pacemark 4410N Wide Format Dot Matrix Printer'), ['Product Name', 'Profit']]
returns_df
# df
```

```python
cleaned_df = df.loc[:, ['Sales', 'Profit', 'Product Category', 'Product Sub-Category', 'Product Name']]

cleaned_df[['Profit']] = cleaned_df[['Profit']].applymap(lambda x: str(x).replace(',',  '.'))
cleaned_df[['Profit']] = cleaned_df[['Profit']].apply(pd.to_numeric)

product_loss_df = cleaned_df.loc[:, ['Product Name', 'Profit']]
product_loss_df = product_loss_df.groupby('Product Name').sum()

category_loss_df = cleaned_df.loc[:, ['Product Sub-Category', 'Profit']]
category_loss_df = category_loss_df.groupby('Product Sub-Category').sum()
```

```python
from datetime import datetime

cleaned_df = df.loc[:, ['Order Date', 'Ship Date', 'Product Category', 'Product Sub-Category', 'Product Name']]

def duration(order_date, ship_date):
    delta = ship_date - order_date
    return delta.days

def convert_date(date):
    return datetime.strptime(date, '%d/%m/%y').date()

cleaned_df['Ship Date'] = cleaned_df['Ship Date'].apply(convert_date)
cleaned_df['Order Date'] = cleaned_df['Order Date'].apply(convert_date)

cleaned_df['Shipping Duration'] = cleaned_df.apply(lambda x: duration(x['Order Date'], x['Ship Date']), axis=1)

sorted_df = cleaned_df.sort_values('Shipping Duration', ascending=False, ignore_index=True)

sorted_df[:20]
```

```python
cleaned_df = df.loc[:, ['Order ID', 'Product Category', 'Product Sub-Category', 'Product Name']]

merged_df = pd.merge(cleaned_df, returns_df, on=['Order ID'])

merged_df = merged_df.groupby('Product Name')[['Status']].count()

sorted_df = merged_df.sort_values('Status', ascending=False)

sorted_df[:20]
# returned_df
```
