```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(16,9)}, font_scale = 1.5)
```

```python
df = pd.read_csv("SuperstoreSales_main.csv", encoding = 'ISO- 8859-1', delimiter = ";")
returns_df = pd.read_csv("SuperstoreSales_returns.csv", encoding = 'ISO- 8859-1', delimiter = ";")

# returns_df
df
```

### Business questions

For this new warehouse in Canada, Mr. Bezos wants to know answers for the following questions

1. Which products/product categories made the **most loss**?
2. Which products/product categories were **shipped really late** (more than 2 days)?
3. Which products/product categories were **returned the most**?


```python
# Clean up columns
cleaned_df = df.loc[:, ['Sales', 'Profit', 'Product Category', 'Product Sub-Category', 'Product Name']]

# Convert profit data from strings to values
cleaned_df[['Profit']] = cleaned_df[['Profit']].applymap(lambda x: str(x).replace(',',  '.'))
cleaned_df[['Profit']] = cleaned_df[['Profit']].apply(pd.to_numeric)

# Sum the total loss per products
product_loss_df = cleaned_df.loc[:, ['Product Name', 'Profit']]
product_loss_df = product_loss_df.groupby('Product Name').sum()

# Sum the total lost per product sub-category
category_loss_df = cleaned_df.loc[:, ['Product Sub-Category', 'Profit']]
category_loss_df = category_loss_df.groupby('Product Sub-Category').sum()
```

#### 1. Which products made the most loss?

The results show the top 20 products that have cost the company the most loss of profit

As observed, there have only been 4 products categories that have cost a loss of profit for the company

```python
# Sort the products by profit and only display the top 20
out = product_loss_df.sort_values('Profit')[:20]
display(out)

sns.barplot(data=out, x='Profit', y=out.index).set(title="Top 20 Products with the most loss")
```

```python
# Sort the product categories by profit and only display the ones that incur a loss
out = category_loss_df.sort_values('Profit')[:4]
display(out)

sns.barplot(data=out, x='Profit', y=out.index).set(title="Top 20 Product sub-categories with the most loss")
```

#### 2. Which products took the longest to be delivered?

The results show the top 20 products that have took the longest to be delivered

As well as the product in each of the 17 product categories that took the longest to be delivered

```python
from datetime import datetime

# Clean up columns
cleaned_df = df.loc[:, ['Order Date', 'Ship Date', 'Product Category', 'Product Sub-Category', 'Product Name']]

# Function to obtain the duration of shipment in number of days
def duration(order_date, ship_date):
    delta = ship_date - order_date
    return delta.days

# Function to convert the date from string to a date type data
def convert_date(date):
    return datetime.strptime(date, '%d/%m/%y').date()

# Convert the Ship Date and Order Date Using the written function
cleaned_df['Ship Date'] = cleaned_df['Ship Date'].apply(convert_date)
cleaned_df['Order Date'] = cleaned_df['Order Date'].apply(convert_date)

# Find the shipping duration using the written function
cleaned_df['Shipping Duration'] = cleaned_df.apply(lambda x: duration(x['Order Date'], x['Ship Date']), axis=1)

# Sort the values based on the shipping duration
sorted_df = cleaned_df.sort_values('Shipping Duration', ascending=False, ignore_index=True)

# Find the product categories with the longest shipping duration
category_shipping_df = sorted_df.drop_duplicates(subset='Product Sub-Category')
```

```python
out = sorted_df[:20]
display(out)
sns.barplot(data=out, x='Shipping Duration', y='Product Name').set(title="Most Delayed Shipments by Product")
```

```python
display(category_shipping_df)
sns.barplot(data=out, x='Shipping Duration', y='Product Name').set(title="Most Delayed Shipments by Product")
```

#### 3. Which products were returned the most?

The results show the top 20 products that have been returned the most

```python
# Clean up columns
cleaned_df = df.loc[:, ['Order ID', 'Product Category', 'Product Sub-Category', 'Product Name']]

# Merge the main dataframe with the returns dataframe based on the order ID
# This will allow us to know if the status of the product has been returned or has no value
merged_df = pd.merge(cleaned_df, returns_df, on=['Order ID'])

# Count the number of times a product has been returned
merged_df = merged_df.groupby('Product Name')[['Status']].count()

# Sort the dataframe by the number of times a product has been returned
sorted_df = merged_df.sort_values('Status', ascending=False)

# Rename Column
sorted_df.columns = ['Returns']

out = sorted_df[:20]
display(out)

sns.barplot(data=out, x='Returns', y=out.index).set(title="Products with the most number of returns")
```
